```python
%matplotlib inline
import matplotlib.pyplot as plt
from PIL import Image, ImageFilter, ImageTransform, ImageOps
```

```python
# visualise given image
image = Image.open('29.ppm')
plt.imshow(image)
```

```python
image = Image.open('29.ppm')
image = image.transform(image.size, Image.AFFINE, (1,0,-4.5,0,1,0))
plt.imshow(image)
```

```python
image = Image.open('29.ppm')
image = image.transform(image.size, Image.AFFINE, (1,0,0,0,1,3.4))
plt.imshow(image)
```

```python
image = Image.open('29.ppm')
image = image.rotate(35)
plt.imshow(image)
```

```python
image = Image.open('29.ppm')
image = image.rotate(-125)
plt.imshow(image)
```

```python
image = Image.open('29.ppm')
image = ImageOps.scale(image, 0.4, resample=3)
plt.imshow(image)
```

```python
image = Image.open('29.ppm')
image = ImageOps.scale(image, 1.4, resample=3)
plt.imshow(image)
```

```python
image = Image.open('29.ppm')
image = ImageOps.equalize(image, mask=None)
plt.imshow(image)
```
